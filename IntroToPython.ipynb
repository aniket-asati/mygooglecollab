```python
# import required libraries

import pandas as pd
```

```python
person = {'First Name':['Surbhi','Aniket'], 'Marks':[100,99]}
person
```

```python
# convert dict to df
df = pd.DataFrame(person)
df
```

```python
# add new columns in df (Techinique 1)
df  = df.assign(City=['New York','Chicago'])
df
```

```python
# add new columns in df (Techinique 2)
df.insert(1,"Section",["A","A"])
df
```

```python
# add new columns in df (Technique 3)
df['Last Name v1'] = ['Nayak','Asati']
df
```
